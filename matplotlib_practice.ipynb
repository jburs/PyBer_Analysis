```python
#import dependencies

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statistics
from matplotlib.ticker import MultipleLocator

# make sure our plots are displayed inline within the Jupyter Notebook
%matplotlib inline
```

```python
savings = [0,100,150,200,300,500,300,300,500,600,750,1000]
months = ["Jan", "Feb", 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Get the standard deviation of the values in the y-axis.
stdev_savings = statistics.stdev(savings)
stdev = statistics.stdev(y_axis)
stdev
```

```python
#Basic line plot with labels

plt.plot(months, savings, color='b')
plt.xlabel('Month')
plt.ylabel('Savings ($)')
plt.title('Saving Account Results')
plt.errorbar(months, savings, yerr=stdev_savings, capsize=3)
plt.show()
```

```python
#same with Object-Oriented inteface, ax.plt()
# e plt.subplots() function returns a tuple that contains a figure and axes object(s).
# fig, ax = plt.subplots() unpacks the tuple into the variables fig and ax.
# The first variable, fig, references the figure.
# The second variable, ax, references the axes of the figure and allows us to annotate the figure.

fig, ax = plt.subplots()
ax.plot(months, savings)
```

```python
# Create the plot and add a label for the legend.
plt.plot(months, y_axis, marker='*', color = "blue", linewidth=3, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid(alpha=.4)
# Add the legend.
plt.legend()
#add error bars
plt.errorbar(months, y_axis, yerr=stdev, capsize = 3)

plt.show()
```

```python
# Repeat above with object oriented method

fix, ax = plt.subplots()
ax.plot(months, y_axis, marker = '*', color ='green', linewidth=1.5, label='Boston')
ax.set_xlim("Jan", "Nov")
ax.set_ylim(0, 46)
ax.set_xlabel('Date')
ax.set_ylabel('Fares ($)')
ax.set_title("PyBer Fare by Month")
ax.grid(alpha = .5)
ax.errorbar(months, y_axis, yerr=stdev, capsize=3)
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot.  #plt.barh() for horizontal bar plot
plt.bar(x_axis, y_axis, color="green", label='Boston', yerr=stdev, capsize=3)
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.  gca = get current axit
plt.barh(x_axis, y_axis, color="magenta", label='Boston', xerr=stdev, capsize=3)
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.xlabel("Fare ($)")
plt.ylabel("Date")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
# specify ticks
plt.xticks(np.arange(0,51, step = 5.0))


```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the horizontal plot with ax.plt()
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis, color='cyan', label='Chicago', xerr=stdev, capsize = 3)
ax.invert_yaxis()
ax.set_ylabel('Month')
ax.set_xlabel('Fare ($)')
ax.set_title("PyBer Fare by Month")
#set ticks
ax.set_xticks(np.arange(0,51, step = 5.0))
#set minor ticks at an increment of 1
ax.xaxis.set_minor_locator(MultipleLocator(1))
ax.legend()
```

```python
#plt.plot(x_axis, y_axis, 'o')
plt.scatter(x_axis, y_axis, color ='red', s=[(.5*i*i) for i in y_axis], label='Chicago')
plt.xlabel("Date")
plt.ylabel("Fare ($)")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color ='royalblue', s=[(.5*i*i) for i in y_axis], linewidth = 2, edgecolors = 'r', alpha = .2, label='Boston')
ax.set_ylabel('Date')
ax.set_xlabel('Fare ($)')
ax.set_title('PyBer Fare by Month')
ax.invert_yaxis()
ax.legend()
```

```python
#pie Chart, explode moves out from 0 of readius
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

plt.pie(y_axis, explode=explode_values, colors=colors, labels = x_axis, autopct='%.1f%%')

plt.show
```
